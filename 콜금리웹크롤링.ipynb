```python
import urllib
import time
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
date_list = []
num_list = []

for i in range(1,62):
    page = 'https://finance.naver.com/marketindex/interestDailyQuote.nhn?marketindexCd=IRR_CALL&page='+str(i)
    html = urlopen(page) 
    source = BeautifulSoup(html.read(), "html.parser")
    a = source.find_all('td', class_='date')
    b = source.find_all('td', class_='num')

    
    for i in range(7):
        date = a[i].text.replace('\n', "").replace('\t', '')
        date_list.append(date)
    
    for i in range(2, 21, 3):
        num_list.append(b[i].text.replace('\n', "").replace('\t', ''))

a = pd.DataFrame({"date" : date_list, "ratio" : num_list})
```
