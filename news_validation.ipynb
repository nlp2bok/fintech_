```python
import pandas as pd
```

```python
news = pd.read_csv('data/news_validation_result.csv')
```

```python
news.dropna(inplace = True)
```

```python
news = news.astype({"predict": int})
```

```python
correct = 0
incorrect = 0

for i in range(len(news)):
    try:
        if ( news['sentiment'].loc[i] == news['predict'].loc[i] ):
            correct += 1

        else:
            incorrect += 1
    except Exception as ex:
        print("에러 ", ex)
        
acc = ( correct / (correct + incorrect) ) * 100

print("")
print("맞은 개수는 ", correct, " 입니다. ")
print("틀린 개수는 ", incorrect, " 입니다. ")
print("총 문장의 개수는", correct + incorrect, " 입니다. ")
print("정확도는 ", acc, " % 입니다.")
```
