```python
news=pd.read_csv("../LSH/data/info_news_2018.csv")
```

```python
news_df=news.drop_duplicates(['content'], keep='first')
```

```python
for i,row in news_df.iterrows() :
    news_df['content'][i] = news_df['content'][i].strip()
    if i % 100 == 0 :
        print('{}/{} '.format(i, news_df.shape[0]))
```

```python
news_df['date']=news_df['date'].str.replace("승인","")
```

```python
for i,row in news_df.iterrows() :
    news_df['date'][i] = news_df['date'][i][2:12]
    if i % 100 == 0 :
        print('{}/{} '.format(i, news_df.shape[0]))
```
